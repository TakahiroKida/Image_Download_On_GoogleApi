```python
#インポート
import cv2
import os
import numpy as np
from google_images_download import google_images_download
```

```python
#検索ワード
keywords = ["dog","cat","deer"]
#ダウンロード数
limit = 100
#保存ディレクトリ
dir_name = "./images"
```

```python
#ディレクトリがなければ作る
if not os.path.isdir(dir_name):
    os.mkdir(dir_name)
#インスタンス生成
res = google_images_download.googleimagesdownload()
```

```python
for key in keywords:
    res.download({"keywords":key,"limit":limit,"format":"jpg","output_directory":dir_name})
```
